```

time= np.linspace(-10,10,100)
x= 5*time**2+2*time
x1= math.e**(2*time)
l= np.convolve(x,x1)
print(l.shape)
print(x.shape)
print(x1.shape)
```

```
n= np.linspace(-199,199,199)
```

```
plt.plot(n,l,'r')
plt.axvline(0,color="black")
plt.axhline(0,color="black")
plt.grid()
plt.show()
```
